```python
# import libraries
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date']   = pd.to_datetime(df['job_posted_date'])
```

```python
df_us = df[df['job_country'] == 'United States'].copy()

df_us
```

```python
df_us['job_posted_monthly'] = df_us['job_posted_date'].dt.strftime('%B')
df_us
```

```python
df_us_pivot = df_us.pivot_table(index='job_posted_monthly', columns='job_title_short', aggfunc='size').sort_index()
```

```python
df_us_pivot.reset_index(inplace=True)
df_us_pivot['month_no'] = pd.to_datetime(df_us_pivot['job_posted_monthly'], format='%B').dt.month

df_us_pivot.sort_values(by='month_no', inplace=True)
df_us_pivot.set_index('job_posted_monthly', inplace=True)

df_us_pivot.drop(columns='month_no', inplace=True)

df_us_pivot.plot(kind='line')
```

```python
top_3 =df_us['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_us_pivot[top_3].plot(kind='line')
```

```python
df_us_pivot[top_3].plot(kind='line')
plt.title('Monthly Job Postings for Top Data Jobs in the US')
plt.xlabel('2023')
plt.ylabel('Job Count')
plt.legend(title='Job Titles')
plt.show()
```
