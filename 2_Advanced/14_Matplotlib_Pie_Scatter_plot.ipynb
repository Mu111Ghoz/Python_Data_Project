```python
# import libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date']   = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
df['job_work_from_home'].value_counts().plot(kind='pie', startangle=90, autopct='%1.1f%%', colors=["#924410", "#8d0f4e"])
plt.title('Work From Home Status')
plt.ylabel("")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='pie')
plt.title('Job Titles')
plt.ylabel("")
plt.show()
```

```python
df['job_title_short'].value_counts().plot(kind='barh')
plt.title('Job Titles')
plt.gca().invert_yaxis()  # Invert y-axis to have the most frequent job title on top
plt.ylabel("")
plt.show()
```

```python
fig, ax = plt.subplots(1, 3)

dict_column = {
    'job_work_from_home': 'Work From Home',
    'job_no_degree_mention': 'Job Degree Req.',
    'job_health_insurance': 'Health Insurance Offered'
}

for i, (column, title) in enumerate(dict_column.items()):
    ax[i].pie(df[column].value_counts(), labels=df[column].value_counts().index, autopct='%1.1f%%', startangle=90)
    ax[i].set_title(title)

plt.show()
```
