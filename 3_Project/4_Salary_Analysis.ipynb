# How well do jobs and skills pay for Data Analysts?

## Methodology

        1. Evaluate median salary for top 6 data jobs.
        2. Find median salary per skill for Data Analysts.
        3. Visualize for highest paying skills and most demanded skills.

```python
# import libraries
import ast
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt
import seaborn as sns

#Loading Data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleanup
df['job_posted_date']   = pd.to_datetime(df['job_posted_date'])
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)
```

```python
# filter for the job titles and country

df_US = df[df['job_country'] == 'United States'].dropna(subset='salary_year_avg')
```

```python
job_titles = df_US['job_title_short'].value_counts().index[:6].to_list()

job_titles
```

```python
df_US_Top6 = df_US[df_US['job_title_short'].isin(job_titles)]
```

```python
job_order = df_US_Top6.groupby('job_title_short')['salary_year_avg'].median().sort_values(ascending=False).index
job_order
```

```python
sns.boxplot(data=df_US_Top6, x='salary_year_avg', y= 'job_title_short', order=job_order)
sns.set_theme(style="ticks")

plt.title('Salary Distributions in the United States')
plt.xlabel('Yearly Salary (USD)')
plt.ylabel('')
plt.xlim(0, 600000)
ticks_x = plt.FuncFormatter(lambda x, pos: f'${x/1000:,.0f}k')
plt.gca().xaxis.set_major_formatter(ticks_x)
plt.tight_layout()
plt.show()
```

# Investigate Median Salary Vs Skill for Data Analysts

```python
df_DA_US = df[(df['job_title_short'] == 'Data Analyst') & (df['job_country'] == 'United States')].copy()

df_DA_US = df_DA_US.dropna(subset=['salary_year_avg'])
```

```python
df_DA_US = df_DA_US.explode('job_skills')

df_DA_US[['salary_year_avg', 'job_skills']].head(5)
```

```python
df_DA_top_pay = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='median', ascending=False)

df_DA_top_pay = df_DA_top_pay.head(10)

df_DA_top_pay
```

```python
df_DA_skills = df_DA_US.groupby('job_skills')['salary_year_avg'].agg(['count', 'median']).sort_values(by='count', ascending=False)

df_DA_skills = df_DA_skills.head(10).sort_values(by='median', ascending=False)

df_DA_skills
```

```python
fig, ax = plt.subplots(2, 1)

sns.set_theme(style="ticks")

# df_DA_top_pay[::-1].plot(kind='barh', y='median', ax=ax[0], title='Top 10 Highest Paying Skills for Data Analysts in the US', legend=False)
sns.barplot(data=df_DA_top_pay, y=df_DA_top_pay.index, x='median', ax=ax[0], hue= 'median', palette='dark:b_r')
ax[0].legend_.remove()  # Remove legend for the first plot
ax[0].set_title('Top 10 Highest Paid Skills for Data Analysts')


sns.barplot(data=df_DA_skills, y=df_DA_skills.index, x='median', ax=ax[1], hue= 'median', palette='light:b')
ax[1].legend_.remove()  # Remove legend for the first plot
# df_DA_skills[::-1].plot(kind='barh', y='median', ax=ax[1], title='Top 10 Most Common Skills for Data Analysts in the US', legend=False)
ax[1].set_xlim(ax[0].get_xlim())  # Align x-axis limits of both plots
ax[0].set_ylabel('')
ax[1].set_ylabel('')
ax[0].set_xlabel('')
ax[1].set_xlabel('Median Salary (USD)')
ax[0].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].xaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'${int(x/1000)}K'))
ax[1].set_title('Top 10 Most In-Demand Skills for Data Analysts')

plt.tight_layout()
fit_layout = plt.gcf()
fit_layout.set_size_inches(10, 8)
plt.show()
```
